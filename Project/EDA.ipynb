```python
import pandas as pd
import matplotlib.pyplot as plt 
from datasets  import load_dataset 
import ast 
import seaborn as sns
# loading data
dataset= load_dataset("lukebarousse/data_jobs")
df=dataset["train"].to_pandas()

# data cleanup  
df["job_posted_date"]=pd.to_datetime(df["job_posted_date"])
df["job_skills"]=df["job_skills"].apply(lambda x : ast.literal_eval(x) if pd.notna(x) else x )
```

```python
df_DA_US=df[(df["job_title_short"]=="Data Analyst")&(df["job_country"]=="United States")].copy()
```

```python
df_DA_US
```

```python
# first of all, l want to look at the job location 
```

```python
df_plot= df_DA_US["job_location"].value_counts().head(10).to_frame()

df_plot
```

```python
sns.set_theme(style="ticks") # to add a little bir more blurred 
sns.barplot(data=df_plot,x="count", y="job_location",hue="count",palette="dark:b_r",legend=False)
sns.despine()  # to remove frame 
plt.title("count of job location for data analyst in the usa ")
plt.xlabel("number of jobs")
plt.ylabel(" ")
plt.show()
```

```python
df_bool=df_DA_US[["job_work_from_home","job_no_degree_mention","job_health_insurance"]]
df_bool
```

```python
liste_column=["job_work_from_home","job_no_degree_mention","job_health_insurance"]
fig,ax = plt.subplots(1,3)
for i,bool in enumerate(liste_column):
    df_bool[bool].value_counts().plot(kind="pie",ax=ax[i],startangle=90,autopct="%1.1f",title=liste_column[i])
    ax[i].set_ylabel("")
fig.tight_layout()
```

```python
df_plot_company= df_DA_US["company_name"].value_counts().head(10).to_frame()

df_plot_company
```

```python
sns.set_theme(style="ticks") # to add a little bir more blurred 
sns.barplot(data=df_plot_company,x="count", y="company_name",hue="count",palette="dark:b_r",legend=False)
sns.despine()  # to remove frame 
plt.title("count of job company for data analyst in the usa ")
plt.xlabel("number of jobs")
plt.ylabel(" ")
plt.show()
```
